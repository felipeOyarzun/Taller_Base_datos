¿Cuál es el número de medalla por cada país, diferenciando el color de la medalla, en los juegos de verano? Visualizar en un histograma.

```python
DForo=DF.loc[DF['Medal'] == 'Gold']
DForo[::]
DForo.groupby('Country')['Medal'].count().plot(kind='bar')
plot.show()
```

```python
DForo=DF.loc[DF['Medal'] == 'Silver']
DForo.groupby('Country')['Medal'].count().plot(kind='bar',color='g')
plot.show()
```

```python
DFbronze=DF.loc[DF['Medal'] == 'Bronze']
DFbronze.groupby('Country')['Medal'].count().plot(kind='bar', color='b')
DFsilver=DF.loc[DF['Medal'] == 'Silver']
DFsilver.groupby('Country')['Medal'].count().plot(kind='bar',color='g')
DForo=DF.loc[DF['Medal'] == 'Gold']
DForo.groupby('Country')['Medal'].count().plot(kind='bar',color='Y')
plot.show()
```
